Removing outliers using 3 standard deviation 

```python
rnd_lower_limit = encoded_df['R&D Spend'].mean() - 3*encoded_df['R&D Spend'].std()
rnd_upper_limit = encoded_df['R&D Spend'].mean() + 3*encoded_df['R&D Spend'].std()
rnd_outliers = encoded_df[(encoded_df['R&D Spend'] > rnd_upper_limit) | (encoded_df['R&D Spend'] < rnd_lower_limit) ]
rnd_outliers
```

There are no outliers
